````python
# ============================================================================
# WORKFLOW FUNCTIONS
# ============================================================================

def run_full_workflow(resume_text=None):
    """Run all 4 agents"""
    
    if not resume_text:
        resume_text = CHRISTIAN_RESUME
    
    print("üöÄ Starting Full Workflow")
    print("   Searching diverse opportunities across all industries")
    print("="*80 + "\n")
    
    # Agent 1: Jobs
    jobs_result = agent_1_job_researcher()
    
    # Agent 2: Profile
    print("\n" + "="*80)
    profile_result = agent_2_personal_profiler(resume_text)
    
    # Agent 3: Resume
    print("\n" + "="*80)
    resume_result = agent_3_resume_strategist(profile_result)
    
    # Agent 4: Interview
    print("\n" + "="*80)
    interview_result = agent_4_interview_coach(profile_result)
    
    # Summary
    print("\n" + "="*80)
    print("‚úÖ WORKFLOW COMPLETE!")
    print("="*80)
    print("\nAll 4 agents completed successfully!")
    print("Scroll up to review all results.")
    
    return {
        "jobs": jobs_result,
        "profile": profile_result,
        "resume": resume_result,
        "interview": interview_result
    }
````

````python
# ============================================================================
# HELPER FUNCTIONS
# ============================================================================

def extract_job_urls(results):
    """Extract clickable URLs from job search results"""
    
    if not results or 'raw_searches' not in results:
        print("No search results to extract URLs from")
        return []
    
    all_urls = []
    for search in results['raw_searches']:
        urls = re.findall(r'https?://[^\s<>"{}|\\^`\[\]]+', str(search['results']))
        all_urls.extend(urls)
    
    all_urls = list(dict.fromkeys(all_urls))  # Remove duplicates
    
    print("\nüîó EXTRACTED JOB URLs:")
    print("="*80)
    for i, url in enumerate(all_urls, 1):
        print(f"{i}. {url}")
    
    print(f"\n‚úÖ Found {len(all_urls)} unique URLs")
    return all_urls
````

````python
def display_job_matches(results):
    """Display job matches in a clean, readable format"""
    
    if not results or 'claude_analysis' not in results:
        print("No results to display")
        return
    
    analysis = results['claude_analysis']
    
    # Try to parse JSON from the analysis
    import json
    try:
        # Extract JSON from markdown code blocks if present
        if "```json" in analysis:
            json_str = analysis.split("```json")[1].split("```")[0]
        elif "```" in analysis:
            json_str = analysis.split("```")[1].split("```")[0]
        else:
            json_str = analysis
        
        jobs = json.loads(json_str.strip())
        
        print("\n" + "="*80)
        print("üìã JOB MATCHES FOR YOU")
        print("="*80)
        
        # Separate by priority if available
        high_priority = [j for j in jobs if j.get('priority') == 'High' or j.get('match_score', 0) >= 8]
        medium_priority = [j for j in jobs if j.get('priority') == 'Medium' or (j.get('match_score', 0) >= 6 and j.get('match_score', 0) < 8)]
        other_jobs = [j for j in jobs if j not in high_priority and j not in medium_priority]
        
        if high_priority:
            print(f"\nüéØ HIGH PRIORITY MATCHES ({len(high_priority)} jobs)")
            print("-"*80)
            for i, job in enumerate(high_priority, 1):
                print(f"\n{i}. {job['title']} at {job['company']}")
                print(f"   üìç {job['location']}")
                print(f"   üíº Experience: {job.get('experience_years', 'N/A')}")
                print(f"   üéØ Match Score: {job.get('match_score', 'N/A')}/10")
                if 'why_you_match' in job:
                    print(f"   ‚úÖ Why: {job['why_you_match'][:100]}...")
                if 'url' in job and job['url'] != 'N/A':
                    print(f"   üîó {job['url']}")
        
        if medium_priority:
            print(f"\n\nüìå MEDIUM PRIORITY ({len(medium_priority)} jobs)")
            print("-"*80)
            for i, job in enumerate(medium_priority, 1):
                print(f"\n{i}. {job['title']} at {job['company']}")
                print(f"   üìç {job['location']}")
                print(f"   üéØ Match: {job.get('match_score', 'Good match')}")
        
        if other_jobs:
            print(f"\n\nüìã OTHER OPPORTUNITIES ({len(other_jobs)} jobs)")
            print("-"*80)
            for i, job in enumerate(other_jobs, 1):
                print(f"{i}. {job['title']} - {job['company']} ({job['location']})")
        
        print("\n" + "="*80)
        print(f"‚úÖ Total: {len(jobs)} jobs found")
        print("="*80 + "\n")
        
    except Exception as e:
        print(f"Could not parse jobs as JSON: {e}")
        print("\nShowing raw analysis instead:")
        print(analysis)
````
